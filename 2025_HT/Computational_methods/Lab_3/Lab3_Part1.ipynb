<p>
    <b>Test number 3</b><br>
For any (invertible) matrix $A$, the inverse times the matrix itself is equal to the Identity matrix, i.e. $A^{-1}A = A A^{-1} = I$, where $I$ is the identity matrix (ones on the main diagonal and zeros elsewhere).<br>
    Use <code>np.random.rand()</code>, to create a $n \times n$ matrix with random numbers (on the range [0 1]). Let $n=4$ and calculate $A^{-1}A$ and $AA^{-1}$, respectively. Print the results and see if it looks like the identity matrix.
</p>

```python
# Enter your code here
n = 4
A = np.random.rand(n, n)
A_inv = np.linalg.inv(A)
I = A @ A_inv

print(A)
print(A_inv)
print(I)
```

<br>
<p>Check the relative errors in the previous computations, $\frac{|| A^{-1}A - I ||}{||I||}$ and $\frac{|| AA^{-1} - I ||}{||I||}$ (you can use any norm). For the exact identity matrix, use <code>np.identity()</code>.
</p>

```python
# Enter your code here
II = np.identity(n)
print(((A_inv @ A) - II) / (II))
```

<p>
What is the order of magnitude of the relative error? </p>
<p>
When it's many repeated computations (like in the inverse-computations), the error tend to get slightly bigger as the number of repeated computation gets larger. Repeat the computation above, but try out different matrix sizes $n$, for example $n=50$, $n=100$, $n=500$, $n≈1000$.
    </p>

```python
# Enter your code here
n = 500
A = np.random.rand(n, n)
A_inv = np.linalg.inv(A)
I = A @ A_inv

print(A)
print(A_inv)
print(I)

print("----------------")
II = np.identity(n)
print(((A_inv @ A) - II) / (II))
```

<br>
<p>The relative error you got in the test examples above was $\approx 10^{-16}$ (if you did it correctly). This is the magnitude of the relative roundoff error and is called <i>machine epsilon</i>, usually denoted $\epsilon_M$. This error is due to rounding of so-called <i>floation point numbers</i> (the way real numbers are represented in the computer). In the next part of this lab we will look more in detail where it comes from.<br>
Note, this is in the very nature of how real numbers are represented in computers. It is not a bug in Python. You’ll see the same kind of thing in all programming languages when you work with so-called floating-point arithmetic. The underlying reason is that the 'mathematical' number axis has infinitely many numbers, and must be approximated with a finite number axis.
    </p>

<p>In most cases the roundoff error is hidden. If you for example print $a=0.1$ and $b = 0.3$ on the screen, Python only display a limited number of decimals, and the numbers are rounded. The result is that you only see <code>0.1</code> and <code>0.3</code>, respectively. It does not mean that the error is gone, it only means that you can't see it.</p>
<p>The code below print $a$ and $b$, displayed in the standard way and displayed with twenty decimal places. You can see the reason why $0.1+0.1+0.1 \neq 0.3$ in floating point addition.</p>

```python
# Run this code
a = 0.1
b = 0.3
print("a=", a, ", and twenty decimal places:", f"{a:.20f}")
print("a=", b, ", and twenty decimal places:", f"{b:.20f}")
```

<h3>To check if reals are equal</h3>
<p>If we actually would like to check if numbers are equal or not, is that possible? You can for example use a code similar to this:</p>
<pre>
  if a == b:
      print('The numbers are equal')
  else:
      print('The numbers are not equal')
</pre>
<p>However, this would not really work becaues the numbers that should be equal aren't really equal. The <code>==</code> would always result in <code>false</code>.</p>
<p>The usual way to solve this is to check if the relative difference between $a$ and $b$ is less then machine epsilon or a number slightly bigger (for example $100 \cdot \epsilon_M$):
$$|a-b|/|a| < 100 \cdot 10^{-16} \Rightarrow |a - b | < |a| \cdot 100\cdot 10^{-16}.$$<br>
The code would change to
<pre>if abs(a-b) < abs(a)*100*1e-16</pre>
Implement this below and check if it works for $a=0.1+0.1+0.1$ and $b=0.3$.
</p>

```python
# Enter your code here
if abs(a - b) < abs(a) * 100 * 1e-16:
    print(True)
else:
    print(False)
```

<p>(If $b=0$, this won't work (divide by zero), and you need to switch to absolute difference in that case).</p>
<p>In <b>numpy</b> there is a built-in function for this, <code>np.isclose(a,b,rtol,atol)</code>, where <code>rtol</code> is the relative tolerance, and <code>atol</code> is the absolute tolerance. Change your if-statement and include <code>np.isclose</code> (you can use <code>rtol=1e-14</code>, <code>atol=1e-14</code>):</p>

```python
# Enter your code here
if np.isclose(a, b, rtol=1e-14, atol=1e-14):
    print(True)
else:
    print(False)
```

<p>Does it also work for $a=0$ ? Let $a=0$ and $b=0.1+0.1+0.1-0.3$ and run the code again.</p>
<p>As you can see it also works for $a=0$ or $b=0$. As usual in <b>numpy</b>, the function is defined also for array input arguments.
