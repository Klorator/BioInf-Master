<h3>Predefined arrays</h3>
<h4><code>zeros</code>, <code>ones</code> and the identity matrix</h4>
<p>First, you will work with three often used predefined arrays, namely <code>np.zeros</code>, <code>np.ones</code> and <code>np.eye</code>.</p>
<p>
    Try the commands <code>A = np.zeros((3,4))</code> and <code>B = np.ones((4,2))</code>. Also print the two matrices on the screen so you can see what they look like.

```python
# Enter your code here
A = np.zeros((3, 4))
B = np.ones((4, 2))

print(A)
print(B)
```

<p>Obviously, the functions creates arrays with all elements as zeros and ones, respectively. The shape of the array is defined by a tuple, in this case <code>(3,4)</code> and <code>(4,2)</code>, which explains the inner paranthesis. The two examples are 2D-arrays (i.e. matrices).<br><br>
If you want a 1D (vector) with zeros (let's say), you just use one number. Try <code>b = np.zeros(6)</code> and print <code>b</code> on the screen. Also check that it really is a 1D-array (use <code>np.shape</code>).</p>

```python
# Enter your code here
b = np.zeros(6)
print(b)
print(b.shape)
print(b.ndim)
```

<p>

<p>The third numpy-function we will deal with here is <code>np.eye</code>. Try <code>np.eye(4)</code> and print the result on the screen.
    <p>

```python
# Enter your code here
print(np.eye(4))
```

<p>The result is a matrix with ones on the main diagonal and zeros elsewhere. In linear algebra this matrix is called the 'Identity matrix', and it is the equivalent to one (1) in the scalar case. If you multiply any matrix $A$ with the identity, the result is just the matrix $A$ itself.<p>
<p>
    It's possible to move the 1's in the main diagonal to other upper or lower diagonals. Try <code>np.eye(6,k=-1)</code> and <code>np.eye(6,k=2)</code>.

```python
# Enter your code here
print(np.eye(6, k=-1))
print(np.eye(6, k=2))
```

<p>
<h4><code>linspace</code> and <code>arange</code></h4>
    Two other commonly used function for generating arrays are <code>np.arange()</code> and <code>np.linspace</code>. Try the command <code>x = np.arange(1, 20)</code> (and print the result).
</p>

```python
# Enter your code here
x = np.arange(1, 20)
print(x)
```

<p>The result is an array with evenly spaced values on the half-open interval [1 20), in other words the interval include 1 but exclude the last value (=20).<br><br>
Change the command to <code>x = np.arange(1, 20, 2)</code>. What happens?</p>

```python
# Enter your code here
x = np.arange(1, 20, 2)
print(x)
```

<p>The last number ('2') is the step (or increment). Here we have increment 2 between the numbers. The default is increment 1.<p>
<p>Create a vector <tt>years</tt>, from 1960 to 2020, with an incremet of 5 years, i.e. $years = \left( \begin{array}{ccccc} 1960 & 1965 & \ldots & 2015 & 2020 \end{array} \right)$. Print it to see if it is correct.</p>

```python
# Enter your code here
years = np.arange(1960, 2020, 5)
print(years)
```

<p>
A similar function is <code>np.linspace</code>, but instead of deciding the increment, you specify starting point, ending point, and <i>the number elements</i>.<br>
    Try <code>np.linspace(-5,10,20)</code>.
    </p>

```python
# Enter your code here
print(np.linspace(-5, 10, 20))
```

<p>If you count the number of elements you'll find 20 elements, starting with -5 and ending with 10. The increment so to speak becomes what it becomes (you don't specify it, as opposed to <code>np.arange</code>).<br><br>
Both these functions are very often used to create the x-axis for plotting: <code>np.arange</code> when you have specified numbers and increments between the numbers, such as <tt>years</tt> above, and <code>np.linspace</code> when you need a high number of elements to get a smooth graph.
    </p>

<h3>Plotting with matplotlib</h3>
<p>In Python <b>matplotlib</b> is the most important package for plotting. Here, you will just get started with a few plots in <i>matplotlib</i> but you can have a look in the <a href="https://matplotlib.org/stable/gallery/index.html" target="_blank">matplotlib gallery</a> to get a sense of what you can do.</p>
<p>First you will get a few examples:</p>


Plot the function $f(x) = x^2$ on the interval [-4 5] (run the code in the cell):

```python
x = np.arange(-4, 5.1, 1)  # Create x-axis (-4 -3 ... 4 5)
fx = x**2  # Compute the function x^2 in the x-values
# Plot the function
plt.plot(x, fx)
plt.title("f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()  # tells Python we are ready to show the plot
```

<p>If you look at the graph, it's a bit jagged. The reason is that the function values are plotted only in the $x$-values (with straight lines in between), and the the step between the numbers is too high. I cases like this, it makes sense to use <code>np.linspace</code>, as we don't have a specified increment. Copy the code, paste it in the  cell below and replace <code>np.arange</code> with <code>np.linspace</code>. Use e.g. 200 points (elements) in $x$.
    </p>

```python
# Enter your code here
# x = np.arange(-4, 5.1, 1)  # Create x-axis (-4 -3 ... 4 5)
x = np.linspace(-4, 5.1, 200)
fx = x**2  # Compute the function x^2 in the x-values
# Plot the function
plt.plot(x, fx)
plt.title("f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()  # tells Python we are ready to show the plot
```

<p>If you've done it correctly, the graph should be more smooth compared with the first plot.</p>
<p>
Plot the function $f(x)=x^2 \cdot cos(x)$ on the interval [-3 3]
</p>

```python
x = np.linspace(-3, 3, 200)  # Create x-axis, -3 to 3 in 200 points
fx = (x**2) * np.cos(x)  # Compute the function x^2*cos(x) in the x-values
# Plot the function
plt.plot(x, fx)
plt.title("f(x) = x^2 * cos(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

<p>One thing to think about here: the function <code>(x**2)*np.cos(x)</code> is actually multiplication between two vectors. The first vector is <code>x**2</code> and the 2nd is <code>np.cos(x)</code>. In part 3 (next part) of this lab we will see that matrix/vector multiplication can be performed in two different ways. It's either a dot-product, or multiplication element by element. As we for every <tt>x</tt>-value are supposed to calculate the formula, it is in this case elementwise multiplication. The use of <tt>*</tt> rather than <tt>@</tt> in the multiplication leads to elementwise multiplication.<br>
What we do here is actually equivalent to a for-loop of the type
    <pre>
    for i in range(len(x)):
      fx[i] = (x[i]**2)*np.cos(x[i])
    </pre>
</p>
<p>
The use of elementwise multiplication is replacing a for-loop (and is also more efficient than using a for-loop).
</p>

<p>Plot the two functions $f(x)=x^2$ and $cos(x)$ on the interval [-3 3], and in the same plot
    </p>

```python
x = np.linspace(-3, 3, 200)  # Create x-axis, -3 to 3 in 200 points
fx1 = x**2
fx2 = np.cos(x)
# Plot the functions
plt.plot(x, fx1)
plt.plot(x, fx2)
plt.title("f(x) = x^2 and cos(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["x^2", "cos(x)"])
plt.show()
```

<p>
    When we have several graphs in the same plot, use <tt>plt.legend</tt> so we know which graph is which. The syntax is <tt>plt.legend([first plot, 2nd plot])</tt>, i.e the same order as in the <tt>plt.plot</tt>-commands.
    </p>
    <p>
  You can change markers and color on the markers, by adding it in the plot-command. For example <code>plt.plot(x,fx1,'g--')</code> stands for green (<tt>g</tt>) and dashed (<tt>--</tt>), and <code>plt.plot(x,fx2,'y:')</code> stands for yellow (<tt>y</tt>) and dotted (<tt>:</tt>). <br>
    Try that change in the box below (copy the code, paste it in the cell and do the changes).
    </p>

```python
# Enter your code here
x = np.linspace(-3, 3, 200)  # Create x-axis, -3 to 3 in 200 points
fx1 = x**2
fx2 = np.cos(x)
# Plot the functions
plt.plot(x, fx1, "g--")
plt.plot(x, fx2, "y:")
plt.title("f(x) = x^2 and cos(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["x^2", "cos(x)"])
plt.show()
```
